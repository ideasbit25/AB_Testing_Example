```python
import numpy as np

import scipy.stats as stats

#control group data (Blue Button)
control_visitors = 1000
control_visitors_conversion = 200

#experiment group data (Red Button)
experiment_visitors = 1000
experiment_visitors_conversion = 230

#calculate conversion rates 
control_conversion_rate = control_visitors_conversion/control_visitors
experiment_conversion_rate = experiment_visitors_conversion/experiment_visitors

#Perform a two sampled t-test
t_stat, p_value = stats.ttest_ind_from_stats(control_conversion_rate, np.sqrt(control_conversion_rate), control_visitors, experiment_conversion_rate, np.sqrt(experiment_conversion_rate), experiment_visitors)

print(t_stat, p_value)

#Result
if p_value<0.05:
    print("The experimental version is statistically better")
else:
    print("The experimental version is not statistically better")
```
